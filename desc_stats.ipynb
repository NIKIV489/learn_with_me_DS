```python
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np

data = np.array([6, 7, 7, 12, 13, 13, 15, 16, 19, 22])
z = stats.zscore(data)

plt.figure()
plt.plot(data, z, marker='o')
plt.axhline(0)
plt.xlabel("Original Data Values")
plt.ylabel("Z-score")
plt.title("Z-score Visualization")
plt.show()
```

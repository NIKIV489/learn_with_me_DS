```python
import matplotlib.pyplot as plt

region = ["west", "north", "east", "south"]
sales = [23, 27, 22, 29]

plt.pie(sales, labels=region, autopct='%1.1f%%',colors=["#ff0000","black","tomato","skyblue"])
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

region = ["west", "north", "east", "south"]
sales = [23, 27, 22, 29]

cmap = plt.cm.viridis
colors = cmap(np.linspace(0, 1, len(sales)))

plt.pie(
    sales,
    labels=region,
    autopct='%1.1f%%',
    colors=colors
)
plt.legend()
plt.show()

```
